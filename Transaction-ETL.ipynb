### Converting the Atm dataframe into sql

```python
df = dim_location
create_table_sql = "CREATE TABLE location (\n"
for column in df.columns:
    data_type = df[column].dtype
    if data_type == 'object':
        data_type = 'VARCHAR(255)'  # Adjust the size based on your data
    elif 'int' in str(data_type):
        data_type = 'INT'
    elif 'float' in str(data_type):
        data_type = 'DECIMAL(10,3)'
    elif data_type == 'datetime64[ns]':
        data_type = 'TIMESTAMP'
    create_table_sql += f"    {column} {data_type},\n"
create_table_sql = create_table_sql.rstrip(',\n')  # Remove the trailing comma and newline
create_table_sql += "\n);"

# Convert DataFrame data to SQL INSERT INTO statements
insert_sql = "INSERT INTO location VALUES\n"
for index, row in df.iterrows():
    values = [f"'{value}'" if isinstance(value, (str, pd.Timestamp)) else str(value) for value in row]
    row_sql = "(" + ", ".join(values) + "),\n"
    insert_sql += row_sql
insert_sql = insert_sql.rstrip(',\n')  # Remove the trailing comma and newline

# Write SQL statements to a SQL file
with open('location.sql', 'w') as f:
    f.write(create_table_sql + '\n\n')
    f.write(insert_sql + ';')
```

### Converting the Transactions dataframe into sql

```python
df = fact_transactions
create_table_sql = "CREATE TABLE transactions (\n"
for column in df.columns:
    data_type = df[column].dtype
    if data_type == 'object':
        data_type = 'VARCHAR(255)'  # Adjust the size based on your data
    elif 'int' in str(data_type):
        data_type = 'INT'
    elif 'float' in str(data_type):
        data_type = 'DECIMAL(10,3)'
    elif data_type == 'datetime64[ns]':
        data_type = 'TIMESTAMP'
    create_table_sql += f"    {column} {data_type},\n"
create_table_sql = create_table_sql.rstrip(',\n')  # Remove the trailing comma and newline
create_table_sql += "\n);"


# Function to write SQL files in chunks
def write_sql_chunks(df, chunk_size):
    total_rows = len(df)
    num_files = total_rows // chunk_size + (1 if total_rows % chunk_size != 0 else 0)
    
    for i in range(num_files):
        start_idx = i * chunk_size
        end_idx = min((i + 1) * chunk_size, total_rows)
        chunk_df = df.iloc[start_idx:end_idx]
        
        insert_sql = "INSERT INTO transactions VALUES\n"
        for _, row in chunk_df.iterrows():
            values = []
            for value in row:
                if pd.isna(value):
                    values.append("NULL")  # Replace NaN with NULL
                elif isinstance(value, (str, pd.Timestamp)):
                    values.append(f"'{value}'")
                else:
                    values.append(str(value))
            row_sql = "(" + ", ".join(values) + "),\n"
            insert_sql += row_sql
        insert_sql = insert_sql.rstrip(',\n')  # Remove the trailing comma and newline
        
        # Write to file
        file_name = f'transactions_{i + 1}.sql'
        with open(file_name, 'w') as f:
            if i == 0:
                f.write(create_table_sql + '\n\n')  # Only include CREATE TABLE in the first file
            f.write(insert_sql + ';')
            
write_sql_chunks(df, 250000)
```
